# Train deepGLM for DirectMarketing Data
### Load DirectMarketing dataset. The data has following information:
• Training data: 900 observations

• Testing data: 100 observations

• Number of covariates: 10

```python
from scipy.io import loadmat
data = loadmat("Sample Code/abaloneBART.mat")
```

## Training Phase
Now start training deepGLM with some settings. By default, if 'distribution' option is not specified then
deepGLMfit will treat y as continuous responses

```python
from deepGLM import toVector, deepGLMfit, deepGLMpredict, deepGLMplot
from scipy.io import loadmat
from numpy import mean, power

data = loadmat("Sample Code/abaloneBART.mat")
X = data['X']
X_test = data['X_test']
y = toVector(data['y'])
y_test = toVector(data['y_test'])
nn = [5,5]

mdl = deepGLMfit(X,y,  
                 network=nn, 
                 lrate=0.01,           
                 verbose=10,             # Display training result each iteration
                 batchsize=X.shape[0],        # Use entire training data as mini-batch
                 epoch=10000,
                 patience=50,          # Higher patience values could lead to overfitting
                 seed=100,
                 windowSize=100)
```

Now plot convergence curve of lowerbound during training phase.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(15,8))
plt.plot(mdl.out.lbBar)
plt.title('Lowerbound of Variational Approximation')
plt.xlabel('Iterations')
plt.ylabel('Lowerbound')
plt.grid()
```

### Variable selection

Now plot Shrinkage coefficients to show if adatptive group Lasso can perform variable selection
efficiently.

Specify shrinkage option to plot shrinkage coefficient from training phase

```python
deepGLMplot('Shrinkage',mdl.out.shrinkage,linewidth=2);
```

### Prediction

The next step is to make prediction from trained model. deepGLM provide point and interval estimation
for unseen data.

In order to make point estimation for new data, use deepGLMpredict function. deepGLMpredict outputs an array of point estimations of observations
in test data

```python
## Prediction on test data
# Make prediction (point estimation) on a test set
print('---------- Prediction ----------')
Pred1 = deepGLMpredict(mdl,X_test);
```

If responses for test data (ytest) is specified (for model evaluation purpose) then we can check PPS
and MSE on test set. Specify test data with option 'ytest'. The output of deepGLMpredict is a structure
including:
    
• An array of prediction values for each observation in test data

• PSS loss of test data

• MSE of test data

```python
# If ytest is specified (for model evaluation purpose)
# then we can check PPS and MSE on test set
Pred2 = deepGLMpredict(mdl,X_test, y_test);
print('PPS on test set using deepGLM is: %s' % Pred2.pps)
print('MSE on test set using deepGLM is: %s' % Pred2.mse)
```

deepGLM also provides prediction interval option.

In order to make interval estimation, specify the interval (within how many std) for option 'Interval' in the
function deepGLMpredict.

Provide sampe size using 'Nsample' options

```python
# Estimate prediction interval for entire test data
Pred4 = deepGLMpredict(mdl,X_test, y=y_test, alpha=1, Nsample=1000)                       
y_pred = toVector(mean(Pred4.yhatMatrix, axis=0))
mse2 = toVector(mean(power(y_test-y_pred,2), axis=0))
accuracy = ((y_test<toVector(Pred4.interval[:,1])) & (y_test>toVector(Pred4.interval[:,0])))
print('Prediction Interval accuracy: %s' % (sum(accuracy)/len(accuracy)))
```

Now we can plot the prediction interval for a subset of unseen data points. Specify 'interval' option
for deepGLMplot

```python
deepGLMplot('Interval', Pred4, y=y_test);
```
